```python
import pandas as pd

import numpy as np


# ── Step 1: Load model-level summaries ───────────────────────────────

sa = pd.read_csv('../layer3/outputs/output_scenario_a.csv')

sb = pd.read_csv('../layer3/outputs/output_scenario_b.csv')

sc = pd.read_csv('../layer3/outputs/output_scenario_c.csv')


# ── Step 2: Apply FIX 4 — frame adoption as GAIN from week 0 baseline

# Adoption starts at ~0.40 (not 0.0). All presentation shows the change.

for df in [sa, sb, sc]:

    df['adoption_gain']   = df['adoption_mean'] - df['adoption_mean'].iloc[0]

    df['gain_p05']        = df['adoption_p05']  - df['adoption_mean'].iloc[0]

    df['gain_p95']        = df['adoption_p95']  - df['adoption_mean'].iloc[0]


# ── Step 3: Verify productivity_delta is a real delta (FIX 1 applied)

# If the fix is applied, values at week 5-8 should be around -0.02 to -0.04

# If NOT applied, values will be ~0.65 — catch this before proceeding

for name, df in [('A', sa), ('B', sb), ('C', sc)]:

    p_vals = df['productivity_delta_true'].values

    if p_vals.mean() > 0.5:

        print(f'ERROR Scenario {name}: productivity_delta looks like absolute values!')

        print('FIX 1 was not applied in run.py. Fix before continuing.')

        break

    else:

        print(f'Scenario {name}: productivity_delta OK. Range: {p_vals.min():.3f} to {p_vals.max():.3f}')


# ── Step 4: Load agent-level data (FIX 2 required) ───────────────────

agents_a = pd.read_parquet('../layer3/outputs/agents_a.parquet')

agents_b = pd.read_parquet('../layer3/outputs/agents_b.parquet')

agents_c = pd.read_parquet('../layer3/outputs/agents_c.parquet')


print(f'Agent data loaded: {len(agents_a):,} rows per scenario')

print('Layer 4 foundation ready.')

```
